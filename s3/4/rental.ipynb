```python
import pandas as pd
```

```python
Adel = pd.read_excel('datasets/AdelaideFinal.xlsx')
QLD = pd.read_excel('datasets/FinalQld.xlsx')
NSW = pd.read_excel('datasets/NSW Final.xlsx')
VIC = pd.read_excel('datasets/VICUpdated.xlsx')
```

```python
print('Adel', Adel.duplicated().value_counts())
print('QLD', QLD.duplicated().value_counts())
print('NSW', NSW.duplicated().value_counts())
print('VIC', VIC.duplicated().value_counts())
```

```python
QLD.drop_duplicates(keep='first', inplace=True)
```

```python
QLD.duplicated().value_counts()
```

```python
cols = ['Postcode', 'Suburb', 'City', 'State', 'Statecode', '1 BR Flat', '2 BR Flat', '3 BR Flat', '4 BR Flat', '1 BR House', '2 BR House', '3 BR House', '4 BR House']
```

```python
Adel = Adel.reindex(columns = cols)
QLD = QLD.reindex(columns = cols)
NSW = NSW.reindex(columns = cols)
VIC = VIC.reindex(columns = cols)
```

```python
rental = pd.concat([Adel, QLD, NSW, VIC])
rental['id'] = range(1, len(rental)+1)
```

```python
rental.to_csv('datasets/finished/rental.csv', index=False)
```
